```python
import pandas as pd
from os import listdir, popen
from os.path import isfile, join, basename,splitext
import pinyin
import numpy as np
%matplotlib inline
```

```python
#遍历目录
dataPath = "../Competition/赛题1数据集"
files = [f for f in listdir(dataPath) if isfile(join(dataPath, f))]
```

```python
#读数据
allData = dict()
for f in files:
    fullPath = join(dataPath, f)
    print(fullPath)
    allData[splitext(f)[0]] = pd.read_excel(fullPath)
```

```python
powers = {'万': 10 ** 4, '亿': 10 ** 8, '%': 0.01}
# add some more to powers as necessary

def removeUnit1(s):
    try:
        power = s[-1]
        if power in powers:
            return float(s[:-1]) * powers[power]
        else:
            return float(s)
    except (TypeError,ValueError):
        return s
    
def removeUnit(x):
    return x.applymap(removeUnit1)

def cleanData(x):
    x.replace(r'^[-]+.*$', np.nan, regex=True,inplace=True)
    return removeUnit(x)
```

```python
#训练数据的标注
score = allData['企业评分']
score.to_excel("real.xlsx", header=False, index=False, engine='xlsxwriter', float_format='%d', sheet_name='Sheet')
```

```python
#专利数量
zl = allData['专利']
zl = cleanData(zl)
zl_count = zl.iloc[:,[0,1]].groupby('企业编号').count()
```

```python
#合并到score
score = score.merge(zl_count,how = 'left', on='企业编号')
```

```python
#融资次数，合并到score
rz = allData['融资信息']
rz = cleanData(rz)
rz_count = rz.iloc[:,[0,2]].groupby('企业编号').count()
score = score.merge(rz_count,how = 'left', on='企业编号')
score.head()
```
